# RAG Exploration Notebook

This notebook demonstrates how to use the RAG pipeline for support query understanding.

```python
from src.ingestion import load_data
from src.embeddings import EmbeddingEngine
from src.retriever import VectorStore
from src.llm_pipeline import LLMPipeline

# 1. Load Data
df = load_data('../data/tickets.csv')
df.head()
```

```python
# 2. Initialize Engines
engine = EmbeddingEngine()
sample_emb = engine.get_embeddings(["test"])
store = VectorStore(sample_emb.shape[1])

# 3. Index Data
embs = engine.get_embeddings(df['cleaned_query'].tolist())
store.add_documents(embs, df)

print(f"Indexed {len(df)} tickets.")
```

```python
# 4. Query & Retrieve
query = "I need a refund for my purchase"
results = store.search(engine.get_embeddings([query]), k=2)
results
```
